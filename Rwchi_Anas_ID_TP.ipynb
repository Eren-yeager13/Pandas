# Exercice 2 : Nettoyage de données 

```python
colonnes_manquantes = df_csv.columns[df_csv.isnull().any()].tolist()
print("Colonnes avec des valeurs manquantes :", colonnes_manquantes)
```

```python
print(df_csv.isna().sum())
```

```python
df_filtree = df_csv.copy()
df_filtree = df_filtree.dropna(subset=['ventes','Profit'])
df_filtree
```

```python
moyenne = df_filtree['satisfaction_client'].mean()
print("la moyenne :",moyenne)
df_filtree['satisfaction_client']=df_filtree['satisfaction_client'].fillna(moyenne)
# x = df_csv['satisfaction_client']
print (df_filtree['satisfaction_client'])
```

```python
# we don't have any duplicated row 
print(df_filtree.duplicated())
```

```python
chiffre_affaires = df_filtree.groupby('Mois')['ventes'].sum()
chiffre_affaires
```
